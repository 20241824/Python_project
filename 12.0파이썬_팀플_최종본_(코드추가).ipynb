```python
import pandas as pd                 # 데이터를 다루기 위해 사용, 월천이의 정보를 엑셀 파일에 저장하거나, 엑셀 파일에서 정보를 불러오는 작업을 할 수 있다.
import os                           # 파일이나 폴더를 다룰때 사용됨, 파일이 있는경우에만 열기 위해 사용, 파일이 없을때 발생하는 기능오류를 방지.
from PIL import Image               # 이미지를 다루기 위한 도구 각 회사의 합격자 스펙을 볼때 사용.
import cv2                          # 이미지나 영상 처리할때 사용.
from google.colab.patches import cv2_imshow  # 이미지가 웹브라우저를 걸쳐서 보여주는걸 방지 Google Colab에서 바로 보여주기 위한 함수.
from google.colab import output
output.enable_custom_widget_manager()

#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ


class StudentInfoSystem:             # 클래스 정의: 학생의 정보를 다루는 시스템을 만들기위해 사용됨. 여러가지 기능을 하나로 묶어서 사용할수 있는 틀.
    def __init__(self):              # 클래스가 처음 만들어질때 실행되는 함수, 일종의 준비작업을 담당. 클래스가 만들어질때 자동 초기화 역할.  '__init__' 는 초기화를 뜻함 student_info와 matched_companies를 초기화 시킴.
        self.student_info = {}       # 학생 정보라는 빈 딕셔너리를 만듦. 학생의 정보를 저장하는 데 사용이 됨.
        self.matched_companies = []  # 빈 리스트를 만듦. 관심분야와 관련있는 회사들을 저장하는데 사용. 학생의 관심사에 맞는 회사를 찾았을 때 그 회사의 이름을 저장하기 위해 사용.
        self.certifications_df = pd.read_excel('Team_Project/Certification.xlsx')


#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def get_student_info(self):      # get_student_info라는 함수를 정의하는 부분. 학생의 이름, 전공 등의 정보를 입력받아 저장하는 역할을 함. 여기에서의 self는 클래스의 일부라는 것을 알려줌.
        try:                         # 코드를 안전하게 실행하기 위한 장치.

            # 학생 정보 입력 받기
            name = input("학생의 이름을 입력하세요: ")

            major = input(f"\n {name} 학생이 학교생활 중 배운 전공을 입력하세요 (콤마로 구분하여 여러 개 입력 가능): ")   # 여기에서 "(f"{name}"는 f-string이라는 방식으로, name 변수에 저장된 학생의 이름을 문자열 안에 쉽게 삽입할 수 있게 해줌.
            major_list = [m.strip() for m in major.split(',')] # 전공(major)을 리스트 형태로 정리하는 부분. "split(',')" 콤마를 기준으로 나눈다. strip() 함수는 문자열의 양쪽 끝에 있는 공백을 제거함.
                        # m 은 사용자가 배운 전공을 입력할때 입력 받은 갑을 잠깐 저장해 놓는 변수임. 직역을 해본다면 "major.split(',')로 문자열을 나누고, 그 리스트에서 각 부분 문자열을 m에 할당한 뒤, 그 m에서 양쪽 끝의 공백을 제거하라."


            # 관심 분야 입력 받기
            print("관심 분야를 선택하세요 (콤마로 구분하여 여러 개 입력 가능):") # 말 그대로 사용자에게 관심분야를 받는거임 .
            print("1. 제조 및 생산 관리")
            print("2. 물류 및 공급망 관리")
            print("3. 품질 관리 및 보증")
            print("4. 프로세스 및 시스템 엔지니어링")
            print("5. 경영 컨설팅 및 분석")
            print("6. R&D 및 기술 개발")
            print("7. IT 및 소프트웨어 분야")
            print("8. 서비스 및 금융")
            print("9. 공공 및 정부 기관")
            interest = input(f"{name} 학생의 관심 분야를 입력하세요: ")          # f-string을 사용 .
            interest_list = [i.strip() for i in interest.split(',')]             # i 또한 m과 같이 사용자가 입력한 관심분야를 잠깐 저장해놓는 변수임 만약 사용자가 콤마를 기준으로 2개를 쓴다면 첫번째i, 두번째i 이렇게 저장이 됨 .
                                                                                 # i.strip(): i에 할당된 문자열의 양쪽 끝에 있는 공백을 제거함.

            print(f"\n{name} 학생의 관심 분야에 해당하는 자격증 목록입니다:")
            for interest in interest_list:
                matched_rows = self.certifications_df[self.certifications_df['관심분야'] == interest]
                if not matched_rows.empty:
                    for _, row in matched_rows.iterrows():
                        certifications = row[1:].dropna().tolist()
                        for cert in certifications:
                            print(f"- {cert}")
                else:
                    print(f"- {interest}에 해당하는 자격증 정보가 없습니다.")





            certifications = input(f"{name} 학생이 가지고 있는 자격증을 입력하세요 (콤마로 구분하여 여러 개 입력 가능): ")  # f-string을 사용.
            certification_list = [c.strip() for c in certifications.split(',')]  # c도 사용자가 입력한 자격증을 잠시 저장해놓는 변수 위에 설명과 동일 .

            # 학점 입력 받기
            grade = input(f"{name} 학생의 학점을 입력하세요 (예: 4.0/4.5 등): ") #f-string을 사용.

            # 토익 점수 입력 받기
            toeic_score = input(f"{name} 학생의 토익 점수를 입력하세요: ")       #f-string을 사용.

            #인턴 경영 있는지 받기
            intern_ex = input(f"{name} 학생은 인턴경험이 있으신가요?(예 6개월,1년,): ")


            # 딕셔너리로 학생 정보 저장
            self.student_info = {                                                # self.student_info는 StudentInfoSystem 클래스 안에서 학생의 정보를 저장하기 위한 공간.
                '이름': name,                                                    # Key와 Value를 설정. 만약 사용자가 학점을 3.8이라고 설정했다며면 grade에 3.8이 저장됨.
                '전공': major_list,
                '관심 분야': interest_list,
                '자격증': certification_list,
                '학점': grade,
                '토익 점수': toeic_score,
                '인턴 경험': intern_ex

            }
        except Exception as e:                                                   # try ~ except 구문에서 except는 오류가 발생했을 때 실행할 코드를 지정하는것임. 앞의 try와 지금 expcept는 한쌍임.
            print(f"학생 정보를 입력하는 중 오류가 발생했습니다: {e}")             #try 블록 안에서 오류가 발생하지 않으면 그냥 다음 코드로 넘어가고, 만약 오류가 발생하면 그 오류에 대해 except 블록이 실행됨.
                                                                                 # as e는 발생한 오류를 e라는 변수에 저장해서 사용할 수 있게 해줌.
                                                                                 # Exception은 모든 종류의 오류(예외)를 다루는 기본 오류 타입임. 어떤 종류의 오류가 발생하든지 간에 이 except 블록이 실행되도록.
                                                                                 # except Exception as e을 직역하자면 "발생한 모든 종류의 오류(Exception)를 처리하라(except), 그리고 그 오류를 e라는 이름으로 저장해."


#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def display_student_info(self):                                              # display_student_info이라는 이름의 함수를 정의. self는 StudentInfoSystem 클래스의 일부라는 것을 알려줌.
        try:                                                                     # try는 오류가 발생할 가능성이 있는 코드를 안전하게 실행시키기 위한 구문
                                                                                 # display_student_info(self)라는 함수의 목적은 클래스에 저장된 학생 정보를 화면에 출력하는것.
                        # 학생 정보 출력
            name = self.student_info['이름']                                     # self.student_info는 학생 정보를 저장한 딕셔너리임.
            majors = ", ".join(self.student_info['전공'])                        # []칸에 해당하는 키 값에 해당하는 값을 가져오라는 뜻. 예를 들어 student_info 딕셔너리에 '이름': '철수'라고 저장되어 있다면, name 변수에는 "철수"가 저장됩니다.
            interests = ", ".join(self.student_info['관심 분야'])                # ", "가 있는 이유는 여러가지를 적을수 있기때문에 그것을 구분하기 위해 존재.
            certifications = ", ".join(self.student_info['자격증'])              # 여러 목록들을 리스트 형태로 가져오고, 이를 join() 함수를 이용해 하나의 문자열로 합칩니다.
            grade = self.student_info['학점']
            toeic_score = self.student_info['토익 점수']
            intern_ex = self.student_info['인턴 경험']

            print(f"{name} 학생의 정보:\n")                                      # f"{name} 학생의 정보:\n"은 f-string을 사용해 학생의 이름을 포함한 문자열을 출력시킴. 여기서 \n은 줄바꿈을 의미함.
            print(f"- 배운 전공: {majors}")
            print(f"- 관심 분야: {interests}")
            print(f"- 자격증: {certifications}")
            print(f"- 학점: {grade}")
            print(f"- 토익 점수: {toeic_score}")
            print(f"- 인턴 경험: {intern_ex}")
        except Exception as e:                                                   # except Exception as e
            print(f"학생 정보를 출력하는 중 오류가 발생했습니다: {e}")



#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def save_student_info_to_excel(self, filename='student_info.xlsx'):          # save_student_info_to_excel은 학생의 정보를 엑셀 파일로 정리하는 함수의 이름임. / self는 클래스(StudentInfoSystem)의 일부임을 나타냄.
                                                                                 # 여기서 filename은 엑셀 파일의 이름을 지정하는 매개변수이고 ='student_info.xlsx'는 기본값으로 설정한것임. 사용자가 별도로 파일이름을 지정하지않으면 자동으로 'student_info.xlsx'라는 이름으로 엑셀파일이 저장됨.
        try:
            # 학생 정보 엑셀 파일로 저장
            data = {                                                             # data = {~}는 학생의 정보를 담고있는 딕셔너리를 만드는 역할. 딕셔너리는 데이터를 '키(key): 값(value)' 쌍으로 저장하는 자료 구조임.
                '이름': [self.student_info['이름']],                             # 여기서 []칸의 내용은 엑셀 파일의 열 이름임. []로 감싸는 이유는 리스트 형태로 만들기 위해서이고.
                '배운 전공': [", ".join(self.student_info['전공'])],             # ", ".join()으로 리스트를 하나의 문자열로 합친 후 리스트 형태로 저장합니다.
                '관심 분야': [", ".join(self.student_info['관심 분야'])],
                '자격증':    [", ".join(self.student_info['자격증'])],
                '학점': [self.student_info['학점']],
                '토익 점수': [self.student_info['토익 점수']],
                '인턴 경험': [self.student_info['인턴 경험']]
            }
            df = pd.DataFrame(data)                                               # 이 줄은 pandas 라이브러리의 DataFrame을 사용해서 데이터프레임을 생성하는 부분입니다. / pd.DataFrame(data)는 앞에서 만든 딕셔너리 data를 이용해서 데이터프레임 객체를 만듦.
            df.to_excel(filename, index=False)                                    # 데이터 프레임은 표와 같은 구조로 데이터를 저장할 수 있는 자료구조로, 엑셀 파일로 변환하거나, 데이터를 다룰 때 매우 유용함
        except Exception as e:                                                    # except Exception as e
            print(f"엑셀 파일로 저장하는 중 오류가 발생했습니다: {e}")


#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def match_interest_with_companies(self, excel_file):                          # match_interest_with_companies은 관심 분야와 관련있는 회사를 찾는 역할을 하는 함수의 이름임. / self는 StudentInfoSystem 클래스의 일부임을 나타냄.
                                                                                  # excel_file: 이 함수는 회사 정보를 포함하고 있는 엑셀 파일을 읽어와야 하므로, 엑셀 파일의 경로를 입력으로 받음.
      try:
            # 엑셀 파일에서 관심 분야에 맞는 회사 찾기
            if not os.path.exists(excel_file):                                    # 이 줄은 엑셀파일이 존재하는지 확인하는 코드임. os.path.exists()는 파일이나 폴더가 실제로 존재하는지를 확인하는 함수.
                print(f"\n'{excel_file}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.") # excel_file**은 함수에 전달된 엑셀 파일의 경로 또는 이름임. 해당 경로에 파일이있으면 True, 없으면 False를 반환.
                return                                                                          # f-string을 사용하여 엑셀 파일의 이름 또는 경로를 포함한 메시지를 출력함.
                                                                                              #'{excel_file}' 부분에는 사용자가 입력한 엑셀 파일의 경로나 이름이 들어감. \n은 줄 바꿈을 의미함.
                                                                                              # return은 함수의 실행을 즉시 종료하는 역할임. 만약 여기서 엑셀 파일이 존재하지않으면, 오류 메세지 출력 후 즉시 종료.
                                                                                              # f로 시작하는 문자열 즉 f-string은 문자열 안에 중괄호({})를 사용하여 변수나 표현식을 넣을수 있음. 변수의 값을 문자열에 삽입하고 싶을때 이 방식이 유용함.
                                                                                              # f-string은 문자열 안에 변수의 값을 간편하게 포함할 수 있도록 도와주는 방식임. / 많은 학생들이 GPT돌리고 프린트 뒤에 f가 왜 쓰이는지, 무슨 역할을 하는지 모르는 학생들이 많음.


            excel_data = pd.read_excel(excel_file)                                            # pandas 라이브러리의 read_excel() 함수를 사용해 엑셀 파일을 읽는 코드임. / pd는 pandas라이브러리를 사용할 때 흔히 사용하는 별칭임. 맨위에 나와있음.
            self.matched_companies = []                                                       # 학생의 관심 분야와 일치하는 회사 목록을 저장하기 위한 변수를 빈 리스트로 초기화하는 코드.
                                                                                              # self.matched_companies: StudentInfoSystem 클래스 내부의 속성(attribute)으로, 학생의 관심 분야와 일치하는 회사를 저장할 공간.
                                                                                              #  = []은 빈 리스트로 초기화. 이 리스트는 이후에 엑셀 데이터에서 학생의 관심 분야와 일치하는 회사를 찾을 때, 그 회사를 추가하는 역할을 하게 됨

            if '분야' not in excel_data.columns:                                              # excel_data.columns: 데이터프레임의 열(column) 이름들을 가져오는 속성이다.
                print("\n엑셀 파일에 '분야' 열이 존재하지 않습니다. 파일을 확인해주세요.")
                return

            for interest in self.student_info['관심 분야']:                                   # 학생의 관심분야를 하나씩 반복해서 처리하는 반복문. / 여기서'분야'는 Colab에 excel.xlsx에 있는 '분야'를 가리킴.  이 '분야'에는 예시로 보여주는 관심분야 목록들이 있음.
                matched_rows = excel_data[excel_data['분야'].str.contains(interest, case=False, na=False)] # 엑셀 데이터에서 학생의 관심 분야와 일치하는 회사들이 있는 행(row)을 찾는 작업. / 열의 각 값에서 학생의 관심분야가 포함 되어있는지를 확인함.
                                                                                                            # case=False: 대소문자를 구분하지 않도록 설정합니다. 예를 들어,'IT'와 'it'를 같은 것으로 취급합니다. / na=False: 결측치(빈 데이터)를 False로 처리하여, 결측치가 있는 행은 포함하지 않도록 합니다.
                                                                                                             # excel_data[ ~ str.contains( ~ )]이를 만족하는 행을 필터링해서 matched_rows에 저장함.

                for _, row in matched_rows.iterrows():                                        # 매칭된 행을 하나씩 처리하기 위한 반복문임. matched_rows 는 이전에 학생의 관심 분야와 일치하는 행들을 저장한 데이터프레임이다. / ".iterrows()": 이 함수는 데이터프레임의 각 행(row)을 하나씩 반복하게 해 줌.
                    companies = row[1:].dropna().tolist()                                     # 각 행은 튜플 형태로 반환되며, 첫 번째 요소는 인덱스 번호, 두 번째 요소는 행 데이터(시리즈 형태)이다.
                    self.matched_companies.extend(companies)                                  # for _, row in matched_rows.iterrows(): 반복문을 통해 matched_rows의 각 행을 row에 담아 처리합니다. / 여기서 _는 인덱스 번호를 의미하지만, 이 값은 사용하지 않음. 그렇기 때문에 _로 무시함
                                                                                              # companies = row[1:].dropna().tolist(): 이 줄은 매칭된 행에서 회사 이름만 추출하여 리스트로 만드는 작업.
                                                                                               # row는 데이터프레임의 한 행을 의미하며, 시리즈 형태로 저장.
                                                                                               # .dropna(): 결측치(빈 데이터)를 제거합니다. 즉, 회사 이름이 없는 부분을 제외하고 나머지만 선택.
                                                                                               # .tolist(): 시리즈 형태로 된 데이터를 리스트 형태로 변환한다.
                                                                                               # companies: 매칭된 행에서 추출한 회사 이름들이 저장된 리스트
                                                                                              # self.matched_companies.extend(companies): 이 줄은 매칭된 회사들을 리스트에 추가하는 부분임 / self는 StudentInfoSystem 클래스의 일부임을 나타냄.
                                                                                               # self.matched_companies는 매칭된 회사들을 저장하는 빈 리스트로, 이전에 초기화된 상태이다.
                                                                                               # .extend(companies): companies 리스트에 저장된 모든 회사 이름을 self.matched_companies 리스트에 추가한다. 예를 들어, companies가 ['삼성', 'LG']라면, 이 회사 이름들이 self.matched_companies에 추가됩니다.



            self.matched_companies = list(set(self.matched_companies))  # 중복 제거           # self.matched_companies는 이전에 학생의 관심 분야와 일치하는 회사들의 이름을 모두 저장한 리스트.
                                                                                              # set(self.matched_companies): set()은 파이썬에서 제공하는 집합 자료형을 만드는 함수임. / set은 중복된 요소를 허용하지 않음.
                                                                                               # set(self.matched_companies)는 self.matched_companies 리스트를 집합으로 변환하여 중복된 회사이름들을 제거함.
                                                                                               # list(set(self.matched_companies))는 집합(set)을 다시 리스트(list)로 변환하는 부분임. / set(self.matched_companies)로 중복이 제거된 목록을 집합형태로 만들고 list() 함수를 사용해 이를 다시 리스트 형태로 변환함.
                                                                                               # self.matched_companies = list(set(self.matched_companies))에서 self.matched_companies 리스트에서 중복된 회사 이름들을 제거하고, 중복이 제거된 결과를 다시 self.matched_companies에 저장하는 역할을 함.
                                                                                              # 이 코드를 실행 시키면 리스트에 중복된 회사 이름들이 없어지고, 고유한 회사들만 포함된 상태로 유지됨.

            if self.matched_companies:                                                        # 이 코드는 조건문으로 self.matched_companies에 매칭된 회사들이 있는지 확인하는 부분.
                                                                                               # self.matched_companies: 이전에 학생의 관심 분야와 일치하는 회사들을 저장한 리스트.
                                                                                               # if self.matched_companies 이 조건문은 self.matched_companies 리스트가 비어있지 않을 때(True) 실행이 됨. 즉 매칭된 회사들이 있을경우에만 그 다음 코드들이 실행됨.

                print("\n관심 분야에 해당하는 회사들:")                                       # 매칭된 회사들이 있을 때, 사용자에게 그 회사들을 출력하기 전에 안내 메시지를 표시하는 역할. / \n은 줄 바꿈.
                for company in self.matched_companies:                                        # 매칭된 회사들을 프린트
                    print(f"- {company}")                                                     # f-string을 사용하여 회사 이름을 문자열에 포함.
                self.get_preferred_companies()                                                # 매칭된 회사들중 가고싶은 회사를 선택하도록 유도하는 부분 사용작가 선택을 했다면 그 결과를 self.student_info
            else:
                print("\n관심 분야에 해당하는 회사를 찾을 수 없습니다.")
      except Exception as e:                                                                  # except Exception as e
            print(f"회사를 매칭하는 중 오류가 발생했습니다: {e}")


#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def get_preferred_companies(self):                                                        # get_preferred_companies은 학생이 추천받은 회사들 중 가고 싶은 회사를 입력받는 함수의 이름. / self는 StudentInfoSystem 클래스의 일부임을 나타냄.

        try:
            # 사용자가 가고 싶은 회사들을 입력 받기
            if self.matched_companies:
                preferred_companies = input("\n위 회사들 중 가고 싶은 회사를 입력하세요 (콤마로 구분하여 여러 개 입력 가능): ") # 사용자가 가고싶은 회사를 적는 칸. \n은 줄바꿈
                preferred_companies_list = [p.strip() for p in preferred_companies.split(',')] # 사용자가 입력한 회사 목록을 정리하여 리스트로 만든부분. preferred_companies 는 학생이 입력한 가고 싶은 회사들을 담고 있음.
                self.student_info['희망 회사'] = preferred_companies_list                       #.split(',')는 콤마(,)를 기준으로 입력 문자열을 나눠서 리스트 형태로 만들어준다. 예를 들어, 학생이 "삼성, 네이버"라고 입력했다면, preferred_companies.split(',')는 ['삼성', ' 네이버']를 반환.
            else:                                                                              # [p.strip() for p in preferred_companies.split(',')]를 직역해본다면 preferred_companies.split(',')로 얻은 리스트의 각 요소 p에 대해 p.strip()을 수행한 결과들을 새로운 리스트로 만들어라./ 이렇게 만들어진 공백이 제거된 리스트가 preferred_companies_list에 저장됨.
                self.student_info['희망 회사'] = []                                            # self.student_info['희망 회사'] = preferred_companies_lis는 학생이 가고싶은 회사 목록을 저장하는 부분임. / self.student_info는 학생의 정보를 저장하고있는 딕셔너리.
        except Exception as e:
            print(f"희망 회사를 입력하는 중 오류가 발생했습니다: {e}")                          # ['희망 회사']: self.student_info 딕셔너리에 '희망 회사'라는 키를 사용해 학생이 가고 싶은 회사 목록을 저장.
                                                                                               # preferred_companies_list: 학생이 입력한 가고 싶은 회사 목록을 정리한 리스트.


#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ


    def display_company_images(self, preferred_companies_list):                  # display_company_images는 회사 이미지들을 화면에 표시하는 역할을 하는 함수의 이름임. self는 StudentInfoSystem 클래스의 일부임을 나타냄.
                                                                                 # preferred_companies_list: 학생이 가고 싶은 회사 목록을 나타내는 매개변수(parameter)

        try:
            # 희망 회사에 따라 이미지 파일 열기
            company_image_map = {
                #  제조 및 생산 관리
                '삼성전자' : 'Team_Project/Samsung e.jpg',
                '현대자동차' : 'Team_Project/Hyundai car.jpg',
                'LG전자' : 'Team_Project/LG e.jpg',
                '포스코' : 'Team_Project/Posco.jpg',
                'SK하이닉스' : 'Team_Project/SK hynix.jpg',
                # 물류 및 공급망 관리
                'CJ대한통운' : 'Team_Project/Cjlogistics.jpg',
                '한진' : 'Team_Project/Hanjin.jpg',
                '롯데글로벌로지스' : 'Team_Project/Lotteglogis.jpg',
                '쿠팡' : 'Team_Project/Coupang.jpg',
                '현대글로비스' : 'Team_Project/Glovis.jpg',
                # 품질 관리 및 보증
                '삼성SDI': 'Team_Project/Samsung SDI.jpg',
                '한국표준협회': 'Team_Project/KSA.jpg',
                'SK이노베이션': 'Team_Project/SK innovation.jpg',
                'LG화학': 'Team_Project/LG chem.jpg',
                '현대모비스': 'Team_Project/Mobis.jpg',
                # 프로세스 및 시스템 엔지니어링
                '한화솔루션' : 'Team_Project/Hanwhasolutions.jpg',
                '두산중공업' : 'Team_Project/Doosanenerbility.jpg',
                'GS칼텍스' : 'Team_Project/GS c.jpg',
                '롯데케미칼' : 'Team_Project/LotteChem.jpg',
                'OCI' : 'Team_Project/OCI.jpg',
                # 경영 컨설팅 및 분석
                '데이터 스트림즈' : 'Team_Project/Data.jpg',
                '딜로이트 안진회계법인' : 'Team_Project/Deloitte.jpg',
                'EY한영회계법인' : 'Team_Project/PWC.jpg',
                'KPMG삼정회계법인' : 'Team_Project/KPMG.jpg',
                'KMAC' : 'Team_Project/KMAC.jpg',
                # R&D 및 기술 개발
                '한국전자통신연구원(ETRI)' : 'Team_Project/ETRI.jpg',
                '한국생산기술연구원' : 'Team_Project/KITECH.jpg',
                'LG이노텍' : 'Team_Project/LGinnotek.jpg',
                '삼성디스플레이' : 'Team_Project/Samsung Display.jpg',
                '현대제철' : 'Team_Project/Hyundai Steel.jpg',
                # IT 및 소프트웨어
                '네이버' : 'Team_Project/Naver.jpg',
                '카카오' : 'Team_Project/Kakao.jpg',
                '삼성SDS' : 'Team_Project/Samsung SDS.jpg',
                'LG CNS' : 'Team_Project/LG CNS.jpg',
                'SK C&C' : 'Team_Project/SK C&C.jpg',
                # 서비스 및 금융
                '신한은행' : 'Team_Project/Sinhan.jpg',
                'KB국민은행' : 'Team_Project/KB.jpg',
                '삼성생명' : 'Team_Project/Samsung life.jpg',
                '하나금융그룹' : 'Team_Project/Hana.jpg',
                '우리은행' : 'Team_Project/Woori.jpg',
                # 공공 및 정부 기관
                '한국산업기술진흥원(KIAT)' : 'Team_Project/KIST.jpg',
                '한국에너지공단' : 'Team_Project/Korea E.jpg',
                '한국도로공사' : 'Team_Project/Korea road.jpg',
                '한국철도공사(KORAIL)' : 'Team_Project/KORAIL.jpg',
                '한국수자원공사(K-water)' : 'Team_Project/K-Water.jpg'
            }

            for company in preferred_companies_list:                             # 학생이 희망하는 회사 리스트(preferred_companies_list)를 반복하면서 각 회사에 대한 이미지를 찾기 위한 반복문./ preferred_companies_list: 이 변수에는 학생이 희망하는 회사들이 리스트 형태로 저장되어 있음
                image_path = company_image_map.get(company)                      # 이 줄은 회사 이름에 해당하는 이미지 경로를 찾는 부분. / company_image_map: 이 변수는 회사 이름과 이미지 파일 경로를 매핑한 딕셔너리(dictionary) 입니다. / .get(company) : 딕셔너리의 get() 메서드를 사용해 현재 반복 중인 회사(company)에 해당하는 이미지 경로를 찾는다.
                if image_path and os.path.exists(image_path):                     # get() 메서드는 키에 해당하는 값을 반환합니다. 만약 키가 존재하지 않는 경우, 기본적으로 None을 반환합니다. / 예를 들어, company가 '삼성'이라면, company_image_map.get('삼성')은 'Team_Project/Samsung.jpg와 같은 이미지 경로를 반환합니다.
                                                                                 # company_image_map.get(company)의 결과는 이미지 파일의 경로이며, 이것을 image_path라는 변수에 저장합니다. / 만약 회사의 이름에 해당하는 이미지 경로가 딕셔너리에 없다면, image_path는 None이 됩니다.
                                                                                 # if image_path and os.path.exists(image_path): 이미지 파일이 존재하는지 확인하는 조건문. / os.path.exists(image_path): os.path.exists()는 주어진 파일 경로나 폴더가 실제로 존재하는지 확인하는 함수이다.
                                                                                  # image_path에 해당하는 경로가 실제로 존재하면 True를 반환하고, 그렇지 않으면 False를 반환합니다.

                                                                                 # if image_path and os.path.exists(image_path)이 조건문은 두 가지 조건이 모두 참일 때만 실행이 된다.
                                                                                  #1-1. image_path가 None이 아닐 것: 즉, 회사 이름에 해당하는 이미지 경로가 딕셔너리에서 잘 찾아졌을 경우.
                                                                                  #1-2. os.path.exists(image_path)가 참일 것: 이미지 파일이 실제로 존재할 경우.
                                                                                   # 이 두 조건이 모두 참(True)이어야, 이후의 코드가 실행된다.


                    try:
                        # 이미지 열기 (Colab 환경에서 cv2_imshow 사용)
                        img = cv2.imread(image_path)                             # OpenCV 라이브러리의 cv2.imread() 함수를 사용해 이미지 파일을 읽는 부분. / image_path는 전에 설정한 변수. / cv2.imread(image_path)는 이미지 파일을 읽어서 OpenCV의 이미지 객체로 반환함. / 예를 들어, image_path가 'Team_Project/Samsung.jpg'라면, 이 파일을 읽어서 img에 이미지 데이터를 저장함.
                        cv2_imshow(img)                                           # img 변수는 이미지 데이터를 담고 있으며, 이를 사용해 이미지를 화면에 표시할 수 있음. / cv2_imshow(img): 읽어들인 이미지를 화면에 표시하는 코드. / cv2_imshow():  Google Colab 환경에서 OpenCV의 이미지를 화면에 출력하는 함수.
                    except Exception as e:
                        print(f"이미지를 여는 중 오류가 발생했습니다: {e}")      # f-string을 사용해 오류 정보를 포함한 메시지를 출력. / 예를 들어, 파일이 없거나 파일 경로가 잘못된 경우, e에는 오류 메시지가 담기며, 이 메시지를 사용자에게 출력한다.
                else:                                                            # if image_path and os.path.exists(image_path): 조건이 거짓(False)일 경우 실행되는 코드 블록을 정의합니다.
                    print(f"'{company}'에 대한 이미지 파일을 찾을 수 없습니다.") # 이 줄은 이미지 파일이 없을 때, 사용자에게 안내 메시지를 출력하는 부분임.
        except Exception as e:
            print(f"회사의 이미지를 표시하는 중 오류가 발생했습니다: {e}")



#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def match_interest_with_certifications(self, cert_excel_file):               # 관심분야에 맞는 자격증을 매칭해주는 역할을 한다.
        try:
            # 엑셀 파일에서 관심 분야에 맞는 자격증 찾기
            if not os.path.exists(cert_excel_file):                              # 이 줄은 엑셀파일이 존재하는지 확인하는 코드임. os.path.exists()는 파일이나 폴더가 실제로 존재하는지를 확인하는 함수.
                print(f"\n'{cert_excel_file}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
                return





            cert_data = pd.read_excel(cert_excel_file)                           # pandas 라이브러리의 read_excel() 함수를 사용해 엑셀 파일을 읽는 코드임.
            student_certs = set(self.student_info['자격증'])                     # self.student_info는 학생의 정보를 저장한 딕셔너리. / self.student_info['자격증']은 딕셔너리에서 '자격증'이라는 키(key)에 해당하는 값을 가져옵니다.

            if '관심분야' not in cert_data.columns:                              # Colab에 들어가있는 자격증 엑셀에 '관심분야'의 열이 없을때 조건이 참이 되어 코드가 실행됨.
                print("\n엑셀 파일에 '관심분야' 열이 존재하지 않습니다. 파일을 확인해주세요.")
                return

            for interest in self.student_info['관심 분야']:                      # self.student_info['관심 분야']는 self.student_info는 학생의 정보를 저장한 딕셔너리이며, **'관심 분야'**는 학생이 입력한 관심 분야 목록을 나타냅니다. / for interest in self.student_info['관심 분야']:  관심 분야(예: 'IT 및 소프트웨어', '경영 컨설팅')를 interest라는 변수에 담아 반복문을 실행합니다.
                matched_rows = cert_data[cert_data['관심분야'].str.contains(interest, case=False, na=False)] # 이 줄은 엑셀 데이터(cert_data)에서 학생의 관심 분야와 일치하는 자격증 정보를 찾는 작업을 함. / cert_data는 자격증 정보가 담긴 엑셀 파일을 데이터프레임 형태로 저장한 변수.
                if not matched_rows.empty:                                       # 이 조건문은 매칭된 자격증 정보가 있는지 확인한다. / matched_rows: 이전에 학생의 관심 분야와 일치하는 자격증 정보가 저장된 데이터프레임. / .empty: 데이터프레임이 비어있는지 여부를 확인하는 속성. / 데이터프레임에 행이 없다면 True, 행이 있으면 False를 반환합니다.
                                                                                  # if not matched_rows.empty:매칭된 자격증 정보가 있을 때(즉, 데이터프레임이 비어있지 않을 때)만 그 다음 코드가 실행됩니다.
                    print(f"\n'{interest}'에 필요한 자격증 정보:")               # 필요한 자격증 정보 출력
                    for _, row in matched_rows.iterrows():                       # 매칭된 자격증을 하나씩 처리하기 위한 코드. / matched_rows는 이전에 학생의 관심 분야와 일치하는 자격증 정보가 저장된 데이터프레임 / .iterrows(): 이 함수는 데이터프레임의 각 행(row)을 하나씩 반복. / 반복문을 통해 matched_rows의 각 행을 row에 담아 처리합니다.
                        certifications = row.dropna().tolist()                   # 위에 이미 설명한게 있음.
                        for cert in certifications:                              # 이 줄은 매칭된 자격증 정보 리스트(certifications)의 각 자격증을 반복하여 처리하는 반복문 / 예를 들어, certifications가 ['정보처리기사', '컴퓨터활용능력 1급']이라면, 이 반복문은 각각의 자격증을 cert라는 변수에 담아서 반복합니다.
                            if cert not in student_certs:                        # 이 줄은 학생이 이미 가지고 있는 자격증인지 확인하는 조건문이다. / student_certs는 학생이 가지고 있는 자격증 목록을 중복 없이 저장한 집합(set)입니다. / cert라는 자격증이 학생이 이미 가지고 있는 자격증 목록(student_certs)에 포함되어 있지 않을 때(즉, 학생이 아직 취득하지 않은 자격증일 때)만 다음 코드가 실행됩니다.
                                print(f"- {cert}")                               # 이 줄은 학생이 아직 가지고 있지 않은 자격증을 화면에 출력하는 역할을 합니다. /
                else:
                    print(f"\n'{interest}'에 해당하는 자격증을 찾을 수 없습니다.")
        except Exception as e:
            print(f"자격증을 매칭하는 중 오류가 발생했습니다: {e}")




#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ




    def compare_scores(self):                                                    # compare_scores 학생의 학점과 토익 점수를 특정 회사의 기준과 비교하는 역할을 하는 함수의 이름
        try:
                                                                                 # data = {~}는 학생의 정보를 담고있는 딕셔너리를 만드는 역할. 딕셔너리는 데이터를 '키(key): 값(value)' 쌍으로 저장하는 자료 구조임.

            # 제조 및 생산 관리 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            if '삼성전자' in self.student_info.get('희망 회사', []):             # 이 조건문은 학생이 희망하는 회사 목록에 '삼성전자'가 포함되어 있는지 확인하는 부분이다. / .get('희망 회사', []): get() 메서드를 사용해 딕셔너리에서 '희망 회사'라는 키에 해당하는 값을 가져온다. / '희망 회사'에 삼성전자가 포함되어 있는지 여부를 확인하는 조건이다.
                excel_file = 'Team_Project/Samsung e.xlsx'                       # 이 줄은 학생이 희망하는 회사가 삼성전자일 경우에 사용할 엑셀 파일 경로를 설정하는 부분이다. / 'Team_Project/Samsung e.xlsx': 이 문자열은 엑셀 파일의 경로와 이름을 나타낸다
            elif '현대자동차' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Hyundai car.xlsx'
            elif 'LG전자' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/LG e.xlsx'
            elif '포스코' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Posco.xlsx'
            elif 'SK하이닉스' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/SK hynix.xlsx'

            # 물류 및 공급망 관리 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif 'CJ대한통운' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Cjlogistics.xlsx'
            elif '한진' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Hanjin.xlsx'
            elif '롯데글로벌로지스' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Lotteglogis.xlsx'
            elif '쿠팡' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Coupang.xlsx'
            elif '현대글로비스' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Glovis.xlsx'

            # 품질 관리 및 보증 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            if '삼성SDI' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Samsung SDI.xlsx'
            elif 'LG화학' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/LG chem.xlsx'
            elif '현대모비스' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Mobis.xlsx'
            elif '한국표준협회' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KSA.xlsx'
            elif 'SK이노베이션' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/SK innovation.xlsx'

            # 프로세스 및 시스템 엔지니어링 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif '한화솔루션' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Hanwhasolutions.xlsx'
            elif '두산중공업' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Doosanenerbility.xlsx'
            elif 'GS칼텍스' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/GS c.xlsx'
            elif '롯데케미칼' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/LotteChem.xlsx'
            elif 'OCI' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/OCI.xlsx'

            #경영 컨설팅 및 분석 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif '데이터 스트림즈' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Data.xlsx'
            elif '딜로이트 안진회계법인' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Deloitte.xlsx'
            elif 'EY한영회계법인' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/PWC.xlsx'
            elif 'KPMG삼정회계법인' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KPMG.xlsx'
            elif 'KMAC' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KMAC.xlsx'

            #R&D 및 기술 개발 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif '한국전자통신연구원(ETRI)' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/ETRI.xlsx'
            elif '한국생산기술연구원' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KITECH.xlsx'
            elif 'LG이노텍' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/LGinnotek.xlsx'
            elif '삼성디스플레이' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Samsung Display.xlsx'
            elif '현대제철' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Hyundai Steel.xlsx'

            #IT 및 소프트웨어 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif '네이버' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Naver.xlsx'
            elif '카카오' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Kakao.xlsx'
            elif '삼성SDS' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Samsung SDS.xlsx'
            elif 'LG CNS' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/LG CNS.xlsx'
            elif 'SK C&C' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/SK C&C.xlsx'

            #서비스 및 금융 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif '신한은행' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Sinhan.xlsx'
            elif 'KB국민은행' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KB.xlsx'
            elif '삼성생명' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Samsung life.xlsx'
            elif '하나금융그룹' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Hana.xlsx'
            elif '우리은행' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Woori.xlsx'

            #공공 및 정부 기관 분야 회사 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            elif '한국산업기술진흥원' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KIST.xlsx'
            elif '한국에너지공단' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Korea E.xlsx'
            elif '한국도로공사' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Korea road.xlsx'
            elif '한국철도공사' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KORAIL.xlsx'
            elif '한국수자원공사' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/K-Water.xlsx'

            else:
                return




#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ

            interests = ", ".join(self.student_info['관심 분야'])


            if not os.path.exists(excel_file):                                   # 이 줄은 엑셀파일이 존재하는지 확인하는 코드임. os.path.exists()는 파일이나 폴더가 실제로 존재하는지를 확인하는 함수.
                print(f"\n'{excel_file}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
                return

            excel_data = pd.read_excel(excel_file)                               # pandas 라이브러리의 read_excel() 함수를 사용해 엑셀 파일을 읽는 코드임.

            if '평균 학점' not in excel_data.columns or '평균토익' not in excel_data.columns :  # 이 줄은 조건문으로, 엑셀 데이터가 필요한 두 개의 열을 모두 포함하고 있는지를 확인 / 이 데이터프레임에는 회사의 평균 학점과 평균 토익 점수가 포함되어 있어야 함.
                print("\n엑셀 파일에 '평균 학점' 또는 '평균토익' 열이 존재하지 않습니다. 파일을 확인해주세요.")
                return                                                           # return은 함수의 실행을 즉시 종료하는 역할임

            average_grade = excel_data['평균 학점'].iloc[0]                      # 이 줄은 엑셀 데이터에서 회사의 평균 학점을 가져오는 작업 / .iloc[]는 데이터프레임의 특정 위치에 있는 값을 가져오는 인덱서 / [0]은 첫 번째 행을 의미함.
            average_toeic = excel_data['평균토익'].iloc[0]                       # 이 줄은 엑셀 데이터에서 회사의 평균 토익 점수를 가져오는 작업.
                                                                                 # 엑셀데이터에서 값을 가져오고, 데이터 프레임에서 특정값도 가져와서 각각 average_grade, average_toeic에 저장합니다.



            student_grade = float(self.student_info['학점'].split('/')[0])       # 이 줄은 학생의 학점을 비교 가능한 형태로 변환하는 작업 / 만약 학생이 '3.2/4.5'라는 학점을 입력했다면 self.student_info['학점']은 '3.2/4.5'를 반환합니다.
                                                                                  # .split('/'): " / "를 기준으로 문자열을 나눠서 리스트로 변환합니다. 변환을하면 ['3.2', '4.5']라는 리스트가 됩니다.
                                                                                  # [0]: 나눈 리스트에서 첫 번째 값(즉, 학생의 실제 학점인 '3.2')을 가져옵니다.
                                                                                  # float(): 문자열로 되어 있는 학점을 실수(float) 형태로 변환합니다. / 예를 들어, '3.2'를 float('3.2')로 변환하여 3.2라는 실수로 저장합니다.
                                                                                  # 학생이 입력한 학점에서 실제 학점 부분만 가져와 실수(float) 형태로 변환하여 student_grade에 저장한다.
            student_toeic = int(self.student_info['토익 점수'])

            if student_toeic < average_toeic:
                print(f"\n 토익 점수를 올리세요! (현재 점수: {student_toeic}, 평균 점수: {average_toeic})")
            if '0' in self.student_info.get('인턴 경험', []):
              print("인턴 경험을 늘려보세요!! (현재 인턴경험:0) ")
            if not '0' in self.student_info.get('인턴 경험', []):
              print(f"학생은 '{interests}' (중)에 어떤 분야에서 인턴생활을 하셨나요? ")
              intern_inform = input(f"\n {interests} 중 어떤것인가요? (콤마로 구분하여 여러 개 입력 가능): ")

            if '0' in self.student_info.get('자격증',[]):
              print(f"\n자격증 공부를 좀 더 준비하세요!")
            if not '0' in self.student_info.get('자격증',[]):
              if student_toeic < average_toeic:
                 print(f"토익점수를 먼저 올리고, 그 다음에 자격증 공부를 해봐요!! (현재 점수: {student_toeic}, 평균 점수: {average_toeic}, 현재 자격증 갯수 : 0)")


              certifications = ", ".join(self.student_info['자격증'])





        except Exception as e:
          print(f"학점 및 토익 점수를 비교하는 중 오류가 발생했습니다: {e}")

        while True:
              response = input("대학원이나 학사과정에 관심이 있습니까? (y/n): ").lower()
              if response == 'y':
                print("대학원이나 학사과정에 관심이 있으시군요!")
                break
              elif response == 'n':
                print("대학원이나 학사과정에 한번 도전해보시면 어떨까요?")
                break
              else:
                print("잘못된 입력입니다. y 또는 n을 입력해주세요.")










#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ



    def run(self):                                                                 # run: 이 함수의 이름입니다. 이 함수는 전체 기능을 하나의 흐름으로 실행하는 역할을 합니다.
        self.get_student_info()                                                    # 이 줄은 학생의 기본 정보를 입력받는 함수를 호출합니다.
        self.display_student_info()                                                # 이 줄은 입력받은 학생 정보를 화면에 표시하는 함수입니다.
        self.save_student_info_to_excel()                                          # 이 줄은 학생의 정보를 엑셀 파일로 저장하는 함수입니다.
        self.match_interest_with_companies('Team_Project/excel.xlsx')              # 이 줄은 학생의 관심 분야와 관련된 회사를 엑셀 파일에서 찾아 매칭하는 함수입니다.
        self.display_company_images(self.student_info.get('희망 회사', []))        # 이 줄은 학생이 가고 싶은 회사의 이미지를 화면에 표시하는 함수입니다.
        self.match_interest_with_certifications('Team_Project/Certification.xlsx') # 이 줄은 학생의 관심 분야와 관련된 자격증을 엑셀 파일에서 찾아 추천하는 함수입니다.

        # 평균 학점 및 토익 점수 비교
        self.compare_scores()

# 메인 코드
if __name__ == "__main__":                                                         # 이 줄은 파이썬 스크립트가 직접 실행될 때만 특정 코드를 실행하도록 설정하는 조건문입니다. / 파이썬 스크립트가 실행될 때, 각 스크립트에는 __name__이라는 내장 변수(Built-in variable)가 있습니다. / 스크립트가 직접 실행되면, __name__의 값은 "__main__"이 됩니다.
                                                                                    # 이 파일이 직접 실행될 때만 특정 코드가 실행되도록 보장하기 위해 사용됩니다.

    student_system = StudentInfoSystem()                                           # 이 줄은 StudentInfoSystem 클래스의 인스턴스를 생성하는 코드입니다.

    student_system.run()                                                           # 이 줄을 통해 학생 정보 시스템의 모든 기능을 한꺼번에 실행합니다. / 프로그램이 시작되면 학생 정보를 입력받고, 그 정보를 처리하여 회사 및 자격증 매칭, 이미지 표시 등 필요한 모든 작업을 수행합니다.


```
