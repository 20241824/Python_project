```python
import pandas as pd
import os
from PIL import Image
import cv2
from google.colab.patches import cv2_imshow

class StudentInfoSystem:
    def __init__(self):
        self.student_info = {}
        self.matched_companies = []

    def get_student_info(self):
        try:
            # 학생 정보 입력 받기
            name = input("학생의 이름을 입력하세요: ")
            major = input(f"{name} 학생이 학교생활 중 배운 전공을 입력하세요 (콤마로 구분하여 여러 개 입력 가능): ")
            major_list = [m.strip() for m in major.split(',')]

            # 관심 분야 입력 받기
            print("관심 분야를 선택하세요 (콤마로 구분하여 여러 개 입력 가능):")
            print("1. 제조 및 생산 관리")
            print("2. 물류 및 공급망 관리")
            print("3. 품질 관리 및 보증")
            print("4. 프로세스 및 시스템 엔지니어링")
            print("5. 경영 컨설팅 및 분석")
            print("6. R&D 및 기술 개발")
            print("7. IT 및 소프트웨어 분야")
            print("8. 서비스 및 금융")
            print("9. 공공 및 정부 기관")
            interest = input(f"{name} 학생의 관심 분야를 입력하세요: ")
            interest_list = [i.strip() for i in interest.split(',')]

            certifications = input(f"{name} 학생이 가지고 있는 자격증을 입력하세요 (콤마로 구분하여 여러 개 입력 가능): ")
            certification_list = [c.strip() for c in certifications.split(',')]

            # 학점 입력 받기
            grade = input(f"{name} 학생의 학점을 입력하세요 (예: 4.0/4.5 등): ")

            # 토익 점수 입력 받기
            toeic_score = input(f"{name} 학생의 토익 점수를 입력하세요: ")

            # 딕셔너리로 학생 정보 저장
            self.student_info = {
                '이름': name,
                '전공': major_list,
                '관심 분야': interest_list,
                '자격증': certification_list,
                '학점': grade,
                '토익 점수': toeic_score
            }
        except Exception as e:
            print(f"학생 정보를 입력하는 중 오류가 발생했습니다: {e}")

    def display_student_info(self):
        try:
            # 학생 정보 출력
            name = self.student_info['이름']
            majors = ", ".join(self.student_info['전공'])
            interests = ", ".join(self.student_info['관심 분야'])
            certifications = ", ".join(self.student_info['자격증'])
            grade = self.student_info['학점']
            toeic_score = self.student_info['토익 점수']

            print(f"{name} 학생의 정보:\n")
            print(f"- 배운 전공: {majors}")
            print(f"- 관심 분야: {interests}")
            print(f"- 자격증: {certifications}")
            print(f"- 학점: {grade}")
            print(f"- 토익 점수: {toeic_score}")
        except Exception as e:
            print(f"학생 정보를 출력하는 중 오류가 발생했습니다: {e}")

    def save_student_info_to_excel(self, filename='student_info.xlsx'):
        try:
            # 학생 정보 엑셀 파일로 저장
            data = {
                '이름': [self.student_info['이름']],
                '배운 전공': [", ".join(self.student_info['전공'])],
                '관심 분야': [", ".join(self.student_info['관심 분야'])],
                '자격증':    [", ".join(self.student_info['자격증'])],
                '학점': [self.student_info['학점']],
                '토익 점수': [self.student_info['토익 점수']]
            }
            df = pd.DataFrame(data)

            df.to_excel(filename, index=False)
        except Exception as e:
            print(f"엑셀 파일로 저장하는 중 오류가 발생했습니다: {e}")

    def match_interest_with_companies(self, excel_file):
        try:
            # 엑셀 파일에서 관심 분야에 맞는 회사 찾기
            if not os.path.exists(excel_file):
                print(f"\n'{excel_file}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
                return

            excel_data = pd.read_excel(excel_file)
            self.matched_companies = []

            if '분야' not in excel_data.columns:
                print("\n엑셀 파일에 '분야' 열이 존재하지 않습니다. 파일을 확인해주세요.")
                return

            for interest in self.student_info['관심 분야']:
                matched_rows = excel_data[excel_data['분야'].str.contains(interest, case=False, na=False)]
                for _, row in matched_rows.iterrows():
                    companies = row[1:].dropna().tolist()
                    self.matched_companies.extend(companies)

            self.matched_companies = list(set(self.matched_companies))  # 중복 제거

            if self.matched_companies:
                print("\n관심 분야에 해당하는 회사들:")
                for company in self.matched_companies:
                    print(f"- {company}")
                self.get_preferred_companies()
            else:
                print("\n관심 분야에 해당하는 회사를 찾을 수 없습니다.")
        except Exception as e:
            print(f"회사를 매칭하는 중 오류가 발생했습니다: {e}")

    def get_preferred_companies(self):
        try:
            # 사용자가 가고 싶은 회사들을 입력 받기
            if self.matched_companies:
                preferred_companies = input("\n위 회사들 중 가고 싶은 회사를 입력하세요 (콤마로 구분하여 여러 개 입력 가능): ")
                preferred_companies_list = [p.strip() for p in preferred_companies.split(',')]
                self.student_info['희망 회사'] = preferred_companies_list
            else:
                self.student_info['희망 회사'] = []
        except Exception as e:
            print(f"희망 회사를 입력하는 중 오류가 발생했습니다: {e}")

    def display_company_images(self, preferred_companies_list):
        try:
            # 희망 회사에 따라 이미지 파일 열기
            company_image_map = {
                '삼성SDI': 'Team_Project/Samsung SDI.jpg',
                '한국표준협회': 'Team_Project/KSA.jpg',
                'SK이노베이션': 'Team_Project/SK innovation.jpg',
                'LG화학': 'Team_Project/LG chem.jpg',
                '현대모비스': 'Team_Project/Mobis.jpg'
            }

            for company in preferred_companies_list:
                image_path = company_image_map.get(company)
                if image_path and os.path.exists(image_path):
                    try:
                        # 이미지 열기 (Colab 환경에서 cv2_imshow 사용)
                        img = cv2.imread(image_path)
                        cv2_imshow(img)
                    except Exception as e:
                        print(f"이미지를 여는 중 오류가 발생했습니다: {e}")
                else:
                    print(f"'{company}'에 대한 이미지 파일을 찾을 수 없습니다.")
        except Exception as e:
            print(f"회사의 이미지를 표시하는 중 오류가 발생했습니다: {e}")

    def match_interest_with_certifications(self, cert_excel_file):
        try:
            # 엑셀 파일에서 관심 분야에 맞는 자격증 찾기
            if not os.path.exists(cert_excel_file):
                print(f"\n'{cert_excel_file}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
                return

            cert_data = pd.read_excel(cert_excel_file)
            student_certs = set(self.student_info['자격증'])

            if '관심분야' not in cert_data.columns:
                print("\n엑셀 파일에 '관심분야' 열이 존재하지 않습니다. 파일을 확인해주세요.")
                return

            for interest in self.student_info['관심 분야']:
                matched_rows = cert_data[cert_data['관심분야'].str.contains(interest, case=False, na=False)]
                if not matched_rows.empty:
                    print(f"\n'{interest}'에 필요한 자격증 정보:")
                    for _, row in matched_rows.iterrows():
                        certifications = row.dropna().tolist()
                        for cert in certifications:
                            if cert not in student_certs:
                                print(f"- {cert}")
                else:
                    print(f"\n'{interest}'에 해당하는 자격증을 찾을 수 없습니다.")
        except Exception as e:
            print(f"자격증을 매칭하는 중 오류가 발생했습니다: {e}")

    def compare_scores(self):
        try:
            # 삼성SDI, LG화학, 현대모비스 선택 시 해당 엑셀 파일에서 평균 학점 및 토익 점수를 불러와 사용자 점수와 비교
            if '삼성SDI' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Samsung SDI.xlsx'
            elif 'LG화학' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/LG chem.xlsx'
            elif '현대모비스' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/Mobis.xlsx'
            elif '한국표준협회' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/KSA.xlsx'
            elif 'SK이노베이션' in self.student_info.get('희망 회사', []):
                excel_file = 'Team_Project/SK innovation.xlsx'
            else:
                return

            if not os.path.exists(excel_file):
                print(f"\n'{excel_file}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
                return

            excel_data = pd.read_excel(excel_file)

            if '평균 학점' not in excel_data.columns or '평균토익' not in excel_data.columns:
                print("\n엑셀 파일에 '평균 학점' 또는 '평균토익' 열이 존재하지 않습니다. 파일을 확인해주세요.")
                return

            average_grade = excel_data['평균 학점'].iloc[0]
            average_toeic = excel_data['평균토익'].iloc[0]

            student_grade = float(self.student_info['학점'].split('/')[0])
            student_toeic = int(self.student_info['토익 점수'])



            if student_toeic < average_toeic:
                print(f"토익 점수를 올리세요! (현재 점수: {student_toeic}, 평균 점수: {average_toeic})")
        except Exception as e:
            print(f"학점 및 토익 점수를 비교하는 중 오류가 발생했습니다: {e}")


    def run(self):
        self.get_student_info()
        self.display_student_info()
        self.save_student_info_to_excel()
        self.match_interest_with_companies('Team_Project/excel.xlsx')
        self.display_company_images(self.student_info.get('희망 회사', []))
        self.match_interest_with_certifications('Team_Project/Certification.xlsx')

        # 평균 학점 및 토익 점수 비교
        self.compare_scores()

# 메인 코드
if __name__ == "__main__":
    student_system = StudentInfoSystem()
    student_system.run()
```
